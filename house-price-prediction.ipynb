```python
df = pd.read_csv("./House Price Prediction Dataset.csv")
df.head()
```

```python
df["Location"].unique()
df["Condition"].unique()
df["Garage"].unique()
```

```python
df["Location"] = df["Location"].map(
    {
        "Downtown": 0,
        "Suburban": 1,
        "Urban": 2,
        "Rural": 3,
    }
)

df["Condition"] = df["Condition"].map(
    {
        "Poor": 0,
        "Fair": 1,
        "Good": 2,
        "Excellent": 3,
    }
)

df["Garage"] = df["Garage"].map(
    {
        "No": 0,
        "Yes": 1,
    }
)
```
